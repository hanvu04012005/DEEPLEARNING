```python
#btvn2
#tạo một tensor có gia trị ban đầu là 2.0
#Định nghĩa hàm số và tính gradient
#y = x^3 + 2x^2 + 5x + 1
#hãy tính dy/dx tại giá trị của x
#Dùng phương pháp Gradient Descent với learning rate alpha = 0.1 để cập nhật giá trị x trong 10 vòng lặp
```

```python
import torch

x = torch.tensor(2.0, requires_grad=True)

# learning rate
alpha = 0.1

print("Bắt đầu: x =", x.item())
print("-" * 60)

for i in range(10):
    # định nghĩa hàm y
    y = x**3 + 2*x**2 + 5*x + 1

    #  tính gradient dy/dx bằng backward()
    y.backward()

    # Lấy gradient hiện tại
    grad = x.grad.item()

    # In thông tin vòng lặp
    print(f"Vòng {i+1:02d}: x = {x.item(): .6f} | y = {y.item(): .6f} | dy/dx = {grad: .6f}")

    #  cập nhật x theo Gradient Descent
    # Phải dùng torch.no_grad() để PyTorch không ghi update này vào graph
    with torch.no_grad():
        x -= alpha * x.grad

    #  reset gradient về 0/None để vòng sau không bị cộng dồn
    x.grad = None

print("-" * 60)
print("Kết thúc: x =", x.item())
```

```python
#btvn3
#Tạo một tập dữ liệu giả lập với x là số giờ học( ngẫu nhiên từ 1 - 10)
#và y là số điểm được tính theo công thức y = 3x + 5 + noise
#Với noise là một giá trị ngẫu nhiên nhỏ
#1. Khởi tạo tham số w và b ngẫu nhiên với requires_grad= true
#2. tính MSE
#3. Tính gradient
#4. Cập nhật tham số w và b bằng gradient Descent với Learning rate alpha = 0.01
#5. Lặp lại quá trình trên trong 100 vòng lặp và quan sát sự hội tụ của mô hình
```

```python
import torch

torch.manual_seed(0)  # cố định seed để dễ tái lập kết quả
# tạo dữ liệu: x là số giờ học từ 1–10
x = torch.randint(1, 11, (100, 1)).float()
# noise nhỏ
noise = torch.randn_like(x) * 0.5
# y = 3x + 5 + noise
y = 3 * x + 5 + noise
# khởi tạo w và b ngẫu nhiên, cho phép tính gradient
w = torch.randn(1, requires_grad=True)
b = torch.randn(1, requires_grad=True)
alpha = 0.01   # learning rate
epochs = 100   # số vòng lặp
for epoch in range(epochs):
    y_pred = w * x + b                 # dự đoán
    loss = torch.mean((y_pred - y)**2) # tính MSE
    loss.backward()                    # tính gradient
    with torch.no_grad():              # cập nhật tham số
        w -= alpha * w.grad
        b -= alpha * b.grad
    w.grad.zero_()                     # reset gradient
    b.grad.zero_()
    if (epoch + 1) % 10 == 0:           # in để quan sát hội tụ
        print(f"Epoch {epoch+1}: Loss={loss.item():.4f}, w={w.item():.4f}, b={b.item():.4f}")
print("w ≈", w.item())
print("b ≈", b.item())
```

Ở trường hợp dùng torch.from_numpy(), tensor và mảng NumPy dùng chung vùng nhớ nên khi thay đổi giá trị của mảng NumPy thì giá trị trong tensor cũng thay đổi theo.
Còn với torch.tensor(), PyTorch tạo một bản sao dữ liệu mới nên tensor và mảng NumPy là độc lập, vì vậy thay đổi mảng NumPy không làm ảnh hưởng đến tensor.

```python
import torch

# tạo tensor rỗng 
a = torch.empty(2, 3)
print("Empty tensor:\n", a)

# tạo tensor toàn số 0
b = torch.zeros(2, 3)
print("\nZeros tensor:\n", b)

# tạo tensor toàn số 1
c = torch.ones(2, 3)
print("\nOnes tensor:\n", c)

# tạo tensor ngẫu nhiên
d = torch.rand(2, 3)
print("\nRandom tensor:\n", d)

# reshape bằng view
e = torch.arange(0, 12)
e_view = e.view(3, 4)
print("\nView reshape:\n", e_view)

# reshape bằng view_as
f = torch.zeros(3, 4)
e_view_as = e.view_as(f)
print("\nView as reshape:\n", e_view_as)
```
