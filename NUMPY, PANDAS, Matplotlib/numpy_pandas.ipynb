```python
#### BTVN 1:cho ma trận = [[99,99,99]
#                          [99,99,99]
#                          [99,99,99]]
#hãy tượng tượng  ô 99 trống
#giả sử 1 là x
# giả sử 0 là o

# lấy đầu vào từ phía X và phía O luân phiên
# nếu đầu vào từ ohias X là (0,0)

# ma trận sẽ trở thành 
#[[x,99,99]
# [99,99,99]
# [99,99,99]]

# nếu đầu vào từ phía o cũng là (0,0)
# thông báo nhập lại 
# nếu không thi điền vào ma trận 

# Thử thách : nếu người đầu tiên có 3 phần tử liên tiếp hãy dừng trò chơi 
```

```python
import numpy as np
import pandas as pd
board = np.full((3, 3), 99)
def print_board():
    print(pd.DataFrame(board))
    print()
def check_win(player):
    for i in range(3):
        if np.all(board[i] == player):
            return True
    for j in range(3):
        if np.all(board[:, j] == player):
            return True
    if np.all(np.diag(board) == player):
        return True
    if np.all(np.diag(np.fliplr(board)) == player):
        return True

    return False
player = 1 
while True:
    print_board()
    name = "X" if player == 1 else "O"
    x = int(input(f"Lượt {name}, nhập hàng (0-2): "))
    y = int(input(f"Lượt {name}, nhập cột (0-2): "))

    if board[x, y] != 99:
        print("Ô đã được đánh, nhập lại!")
        continue
    board[x, y] = player

    if check_win(player):
        print_board()
        print(f"{name} thắng!")
        break
    player = 0 if player == 1 else 1
```

```python
import numpy as np
from sklearn.model_selection import train_test_split, KFold
data = np.random.rand(150, 5)

#  Tách mảng thành X (4 cột đầu) và y (cột cuối)
X = data[:, :4]  # Lấy tất cả hàng, từ cột 0 đến cột 3
y = data[:, 4]   # Lấy tất cả hàng, chỉ lấy cột 4

#  Tách X, y thành tập train (70%) và test (30%)
# Sử dụng shuffle=True để dữ liệu được xáo trộn ngẫu nhiên trước khi chia
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Tạo 10 tập dữ liệu không chồng chéo nhau (K-Fold Cross Validation)
kf = KFold(n_splits=10)

print(f"Kích thước X_train: {X_train.shape}")
print(f"Kích thước X_test: {X_test.shape}")
print("-" * 30)
for i, (train_index, val_index) in enumerate(kf.split(X)):
    print(f"Tập thứ {i+1}: Index kiểm tra từ {val_index[0]} đến {val_index[-1]}")
```
